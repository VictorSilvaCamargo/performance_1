```python
import decimal
import timeit
import sys

def calcular_pi():
    # Definir a precisão desejada (101 para imprimir 100 casas decimais)
    decimal.getcontext().prec = 101

    # Calcular o valor de pi utilizando a aproximação de Leibniz
    pi = decimal.Decimal(0)
    k = 0
    while k < 100000: #quanto maior o número do loop, mais preciso será o resultado
        pi += decimal.Decimal((-1) ** k) / (2 * k + 1)
        k += 1
    pi *= 4
    
    return pi

# Definição da variável k antes da chamada da função calcular_pi()
k = 0

# Chamada da função calcular_pi()
pi = calcular_pi()

# Impressão do valor de pi com 100 casas decimais
print(pi)

# Medição do tempo de execução da função calcular_pi()
tempo_execucao = timeit.timeit(lambda: calcular_pi(), number=1)
print(f"Tempo de execução de calcular_pi: {tempo_execucao * 10**9:.2f} ns")

# Medição do espaço de memória utilizado pela função calcular_pi()
memoria_utilizada = sys.getsizeof(decimal.Decimal(0)) + sys.getsizeof(0) + sys.getsizeof(k) + sys.getsizeof(pi)
print(f"Espaço de memória utilizado por calcular_pi: {memoria_utilizada} bytes")
```

```python
import timeit
import random
import sys

# cria um array com 1000 caracteres aleatórios
def criar_array():
    caracteres = []

    for i in range(1000):
        numero_aleatorio = random.randint(0, 10000)
        caracteres.append(numero_aleatorio)

    tamanho_memoria = sys.getsizeof(caracteres)
    return caracteres, tamanho_memoria

# ordena (do menor para o maior) os caracteres do array
def ordenar(caracteres):
    caracteres.sort()

    tamanho_memoria = sys.getsizeof(caracteres)
    return caracteres, tamanho_memoria

# mede o tempo de execução da função criar_array
tempo_criar_array = timeit.timeit(lambda: criar_array(), number=1)

# mede o tempo de execução da função ordenar
caracteres, tamanho_memoria = criar_array()
tempo_ordenar = timeit.timeit(lambda: ordenar(caracteres), number=1)

# exibe o tempo de execução e espaço de memória utilizado de cada função
print(f"Array gerado por criar_array: {criar_array()[0]}")
print(f"Tempo de execução de criar_array: {tempo_criar_array * 10**9:.2f} ns")
print(f"Tamanho da memória utilizada por criar_array: {tamanho_memoria} bytes")
print(f"Array ordenado por ordenar: {ordenar(caracteres)[0]}")
print(f"Tempo de execução de ordenar: {tempo_ordenar * 10**9:.2f} ns")
print(f"Tamanho da memória utilizada por ordenar: {tamanho_memoria} bytes")
```
